```python
import pandas as pd
import numpy as np
from alphien.utils import get_all_data, _read_output
```

```python
def add_features(data):
    data = data.copy()
    data['date'] = pd.to_datetime(data['date'])
    data = data.set_index('date')
    data = data.dropna()
    return data
```

```python
datalist = list(_read_output().values())
datalist = [data for data in datalist if data.shape[0] > 1]
datalist = [add_features(data) for data in datalist]
```

```python
df = pd.concat(datalist)
df = df.reset_index()
df = df.sort_values(by=['date','ticker'])
df = df.set_index(['date','ticker'])
df.head(120)
```

```python
df2 = df.reset_index()
```

```python
labels = pd.read_csv(os.path.join(os.getcwd(), "labels.csv")).drop(["Unnamed: 0"], axis = 1)
```

```python
df3 = df2.merge(labels, left_on= ["date", "ticker"], right_on= ["date", "ticker"], how="left")
```

```python
Target = df_train.target
Predictors = df_train.drop(["date", "ticker", "target"], axis = 1)
```
